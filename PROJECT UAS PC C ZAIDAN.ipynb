```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar
image_path = "NANAS.jpg"
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Mengubah gambar ke HSV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Segmentasi buah nanas 
lower_nanas = np.array([0, 50, 50])  
upper_nanas = np.array([50, 255, 255])  
mask_nanas = cv2.inRange(hsv, lower_nanas, upper_nanas)
segmented_nanas = cv2.bitwise_and(image_rgb, image_rgb, mask=mask_nanas)

# Segmentasi daun nanas berdasarkan rentang warna
lower_green = np.array([25, 40, 40])
upper_green = np.array([95, 255, 255])
mask_daun = cv2.inRange(hsv, lower_green, upper_green)
segmented_daun = cv2.bitwise_and(image_rgb, image_rgb, mask=mask_daun)

# Menampilkan hasil
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title("Gambar Asli")
plt.imshow(image_rgb)
plt.axis('on')

plt.subplot(2, 2, 2)
plt.title("Mask Nanas")
plt.imshow(mask_nanas, cmap='gray')
plt.axis('on')

plt.subplot(2, 2, 3)
plt.title("Segmentasi Nanas")
plt.imshow(segmented_nanas)
plt.axis('on')

plt.subplot(2, 2, 4)
plt.title("Segmentasi Daun")
plt.imshow(segmented_daun)
plt.axis('on')

plt.tight_layout()
plt.show()
```
